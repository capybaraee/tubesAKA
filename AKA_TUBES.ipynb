```python
import pandas as pd

# Sample dataset (replace this with your actual dataset loading)
# Assuming you have already loaded 'df' and selected the first 300 rows:
df_300 = df.head(300)  # Select the first 300 rows from the dataset

# Select only the relevant columns (ProductName, StockQuantity)
df_filtered = df_300[['ProductName', 'StockQuantity']]

# Convert the DataFrame to a list of dictionaries
products = df_filtered.to_dict(orient='records')

# Recursive Insertion Sort function
def insertion_sort_recursive(products, n=None):
    if n is None:
        n = len(products)

    # Kasus dasar: Jika n kurang dari atau sama dengan 1, berarti list sudah terurut
    if n <= 1:
        return

    # Melakukan rekursi untuk menyortir n-1 elemen pertama
    insertion_sort_recursive(products, n - 1)

    # Memasukkan elemen ke-n ke posisi yang benar
    key = products[n - 1]  # Elemen yang akan diposisikan dengan benar
    j = n - 2  # Membandingkan elemen yang lebih kecil (elemen di kiri)

    # Selama j >= 0 dan elemen di kiri lebih besar dari key, geser elemen ke kanan
    while j >= 0 and products[j]['StockQuantity'] > key['StockQuantity']:
        products[j + 1] = products[j]  # Geser elemen ke kanan
        j -= 1  # Pindah ke elemen berikutnya

    # Tempatkan key di posisi yang benar
    products[j + 1] = key


# Call the recursive sorting function
insertion_sort_recursive(products)

# Convert the sorted list of dictionaries back to a DataFrame for tabular output
df_sorted = pd.DataFrame(products)

# Adjust pandas display settings to show all rows
pd.set_option('display.max_rows', None)  # Remove row limit
pd.set_option('display.max_columns', None)  # Remove column limit

# Display the sorted DataFrame as a table
print(df_sorted)
```

```python
import time
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sys

# Increase recursion limit
sys.setrecursionlimit(2000)

# Sample data generation function
def generate_data(size):
    return [{'ProductName': f'Product {i}', 'stok': np.random.randint(1, 500)} for i in range(size)]

# Recursive Insertion Sort function
def insertion_sort_iterative(products):
    # Mulai dari elemen kedua (index 1) karena elemen pertama sudah terurut
    for i in range(1, len(products)):
        # Ambil elemen yang akan disisipkan
        last = products[i]
        j = i - 1

        # Geser elemen-elemen yang lebih besar dari last ke kanan untuk memberi ruang
        while j >= 0 and products[j]['stok'] > last['stok']:
            products[j + 1] = products[j]
            j -= 1

        # Tempatkan last pada posisi yang benar
        products[j + 1] = last

# Recursive Quick Sort function
def quick_sort_recursive(products, low, high):
    if low < high:
        pivot_index = partition(products, low, high)
        quick_sort_recursive(products, low, pivot_index - 1)
        quick_sort_recursive(products, pivot_index + 1, high)

def partition(products, low, high):
    pivot = products[high]['stok']
    i = low - 1
    for j in range(low, high):
        if products[j]['stok'] <= pivot:
            i += 1
            products[i], products[j] = products[j], products[i]
    products[i + 1], products[high] = products[high], products[i + 1]
    return i + 1

# Function to measure execution time for Recursive Insertion Sort
def measure_insertion_sort_time(size):
    data = generate_data(size)
    start_time = time.time()
    insertion_sort_recursive(data)
    end_time = time.time()
    return end_time - start_time

# Function to measure execution time for Recursive Quick Sort
def measure_quick_sort_time(size):
    data = generate_data(size)
    start_time = time.time()
    quick_sort_recursive(data, 0, len(data) - 1)
    end_time = time.time()
    return end_time - start_time

# Sizes of input to test (e.g., 100, 200, ..., 1000)
sizes = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]

# Measure times for both sorting algorithms
insertion_times = [measure_insertion_sort_time(size) for size in sizes]
quick_sort_times = [measure_quick_sort_time(size) for size in sizes]

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(sizes, insertion_times, label='Insertion Sort (Recursive)', marker='o', color='blue')
plt.plot(sizes, quick_sort_times, label='Quick Sort (Recursive)', marker='o', color='red')

# Add text for time values at each point
for i, txt in enumerate(insertion_times):
    plt.text(sizes[i], insertion_times[i], f'{txt:.6f}', fontsize=9, color='blue', ha='right')

for i, txt in enumerate(quick_sort_times):
    plt.text(sizes[i], quick_sort_times[i], f'{txt:.6f}', fontsize=9, color='red', ha='left')

plt.xlabel('Input Size')
plt.ylabel('Execution Time (seconds)')
plt.title('Comparison of Execution Time: Insertion Sort vs Quick Sort (Recursive)')
plt.legend()
plt.grid(True)
plt.show()

# Print the execution times for each input size
print("Input Size | Insertion Sort Time (s) | Quick Sort Time (s)")
for i, size in enumerate(sizes):
    print(f"{size:>10} | {insertion_times[i]:.5f}                 | {quick_sort_times[i]:.5f}")
```

```python
import time
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Sample data generation function
def generate_data(size):
    return [{'ProductName': f'Product {i}', 'stok': np.random.randint(1, 500)} for i in range(size)]

# Iterative Insertion Sort function
def insertion_sort_iterative(products):
    for i in range(1, len(products)):
        key = products[i]
        j = i - 1
        while j >= 0 and products[j]['stok'] > key['stok']:
            products[j + 1] = products[j]
            j -= 1
        products[j + 1] = key

# Iterative Quick Sort function
def quick_sort_iterative(products):
    stack = [(0, len(products) - 1)]
    while stack:
        low, high = stack.pop()
        if low < high:
            pivot_index = partition(products, low, high)
            stack.append((low, pivot_index - 1))
            stack.append((pivot_index + 1, high))

def partition(products, low, high):
    pivot = products[high]['stok']
    i = low - 1
    for j in range(low, high):
        if products[j]['stok'] <= pivot:
            i += 1
            products[i], products[j] = products[j], products[i]
    products[i + 1], products[high] = products[high], products[i + 1]
    return i + 1

# Function to measure execution time for Iterative Insertion Sort
def measure_insertion_sort_time(size):
    data = generate_data(size)
    start_time = time.time()
    insertion_sort_iterative(data)
    end_time = time.time()
    return end_time - start_time

# Function to measure execution time for Iterative Quick Sort
def measure_quick_sort_time(size):
    data = generate_data(size)
    start_time = time.time()
    quick_sort_iterative(data)
    end_time = time.time()
    return end_time - start_time

# Sizes of input to test (e.g., 100, 200, ..., 1000)
sizes = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]

# Measure times for both sorting algorithms
insertion_times = [measure_insertion_sort_time(size) for size in sizes]
quick_sort_times = [measure_quick_sort_time(size) for size in sizes]

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(sizes, insertion_times, label='Insertion Sort (Iterative)', marker='o', color='blue')
plt.plot(sizes, quick_sort_times, label='Quick Sort (Iterative)', marker='o', color='red')

# Add text for time values at each point
for i, txt in enumerate(insertion_times):
    plt.text(sizes[i], insertion_times[i], f'{txt:.6f}', fontsize=9, color='blue', ha='right')

for i, txt in enumerate(quick_sort_times):
    plt.text(sizes[i], quick_sort_times[i], f'{txt:.6f}', fontsize=9, color='red', ha='left')

plt.xlabel('Input Size')
plt.ylabel('Execution Time (seconds)')
plt.title('Comparison of Execution Time: Insertion Sort vs Quick Sort (Iterative)')
plt.legend()
plt.grid(True)
plt.show()

# Print the execution times for each input size
print("Input Size | Insertion Sort Time (s) | Quick Sort Time (s)")
for i, size in enumerate(sizes):
    print(f"{size:>10} | {insertion_times[i]:.5f}                 | {quick_sort_times[i]:.5f}")
```
